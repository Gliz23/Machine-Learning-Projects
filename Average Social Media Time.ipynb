```python
# THe mean hours spent is

mean_time = socialmedia_data.time_spent.mean()
mean_time
```

```python
# To get the standard deviation of the time spent,

# we can use a map function to get the deviation since the new series does 
# not affect the original data. (ts - time spent)

Time_diviation = socialmedia_data.time_spent.map(lambda ts: ts - mean_time)

#Square them (td- time deviation)
Time_diviation_squared = Time_diviation.map(lambda td: td ** 2)

#The standard diviation
Var_Time_Spent = Time_diviation_squared.mean()

SD_Time_Spent = np.sqrt(Var_Time_Spent)
 
SD_Time_Spent
```

```python
#If we want to segregate the data into categories, we use groupby

unique_countires_count = socialmedia_data.groupby('location').location.count()


unique_countires_count
```

```python
#To know the total time spent per country, we can use a function
def ttspent_pcountry(Country):
    #group the countries
    unique_countires = socialmedia_data.groupby('location')

    #Pick the group we need
    Country_data = unique_countires.get_group(Country)

    #Pick the timespent value, sum them and display 
    return Country_data.time_spent.sum()


Australia_data = ttspent_pcountry('Australia') 

Australia_data 
    
    
    
```

```python
#To know the amount of time spent per country in the given data:

unique_countires = socialmedia_data.groupby('location')

country_names = unique_countires.groups.keys()

for i in country_names:
    print(f'{i} : {ttspent_pcountry(i)}')
    
  
```
